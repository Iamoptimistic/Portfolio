```python
#This code remove non alphanumeric character from the year and still create a new column called year of release
movies['YEAR OF RELEASE'] = movies['YEAR']. str.replace('[^a-zA-Z0-9]', '', regex=True)
```

```python
movies
```

```python
# The next step is to drop the YEAR column
movies = movies.drop('YEAR', axis=1)
```

```python
movies
```

```python
#Sorting the year of release to the first time of release by keeping the first four character of the column called YEAR OF RELEASE
movies['YEAR'] = movies['YEAR OF RELEASE'].str[:4]
```

```python
movies
```

```python
#removing the YEAR OF RELEASE column
movies = movies.drop('YEAR OF RELEASE', axis=1)
```

```python
movies
```

```python
# Removing the first two character of the column of GENRE, ONE LINE and STAR
movies['GENRE'] = movies['GENRE'].str[2:]
movies['ONE-LINE'] = movies['ONE-LINE'].str[2:]
movies['STARS'] = movies['STARS'].str[2:]
```

```python
movies
```

```python
'''
I have made a mistake here which is irreversible, there are two ways which might be that i have to reload all the data back again or i have to just make some arrangement to modify it. 
Its always better to create a new column from one than editing it straight.
'''
```

```python
#Let me attempt to save only the last complete words in that GENRE
movies['GENRE OF THE MOVIE'] = movies['GENRE'].str.split().str[-1]
```

```python
movies
```

```python
#Then drop GENRE
movies= movies.drop('GENRE', axis= 1)
```

```python
movies
```

```python
# Let me do the right thing now
movies['STAR'] = movies['STARS'].str.replace('\n', ' ')
movies
```

```python
#Drop STARS
movies = movies.drop('STARS', axis= 1)
movies
```

```python
#Since i have spoilt the data of ONE-LINE, Let me remove it
movies = movies.drop('ONE-LINE', axis= True)
movies
```

```python
movies.head
```

```python
# Now looking out for Replica in the title of the movie
# Let me first use a boolean to know if there is a replica
duplicates = movies.duplicated()
print(duplicates)
```

```python
#Print only where there is duplicate
print(movies[duplicates])
```

```python
#Using Value count for each column
value_count = movies['MOVIES'].value_counts()
duplicates = value_count[value_count > 1]
print(duplicates)
```

```python
#The result above showed that we have enough duplicate values for the movie list
# I will attempt to remove the replica and leave the first for each replica
movies_no_duplicates = movies.drop_duplicates(subset='MOVIES', keep= 'first')
```

```python
#The number is expected to reduce when the replica has been remove.
movies_no_duplicates
```

```python
# Lets check for replica again, we expect a none which is what we see.
value_count = movies_no_duplicates['MOVIES'].value_counts()
duplicates = value_count[value_count > 1]
print(duplicates)
```

```python
movies_no_duplicates
```

```python
# Updating our data with implace
movies= movies.drop_duplicates(subset='MOVIES', keep= 'first')
```

```python
# lets check the number of movies released per year 
#Importing library
import matplotlib.pyplot as plt
```
